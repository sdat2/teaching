```python
import numpy as np
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import gamma
from scipy.stats import uniform
from scipy.stats import beta
from scipy.stats import t
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1)
n, p = 100, 0.4
x = np.arange(binom.ppf(0.001, n, p),
              binom.ppf(0.999, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
a = 0.99
x = np.linspace(gamma.ppf(0.01, a),
                gamma.ppf(0.99, a), 100)
ax.plot(x, gamma.pdf(x, a),
       'r-', #lw=5, 
        alpha=0.6, 
        label='gamma pdf')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(uniform.ppf(0.01),
                uniform.ppf(0.99), 100)
ax.plot(2*x, uniform.pdf(x)/2,
       'r-', lw=5, alpha=0.6, label='uniform pdf')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
df = 2.74
x = np.linspace(t.ppf(0.01, df),
                t.ppf(0.99, df), 100)
ax.plot(x, t.pdf(x, df),
       'r-', lw=5, alpha=0.6, label='t pdf')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
param_list = [[0.5,0.5], [1.5, 1.5], [1, 2], [0.5, 2]]
options = ['a', 'b', 'c', 'd']
for i in range(len(param_list)):
    a, b = param_list[i]
    x = np.linspace(0, 1, 100)
    ax.plot(x, beta.pdf(x, a, b),
           '-', alpha=0.6, label='beta pdf (' + options[i] + ')')
    plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()
```

```python
from scipy import integrate

lamb = 5 # arbitrary number

def a(u):
    return lamb * np.exp(-lamb * np.sqrt(u))

def b(u):
    return lamb/u * np.exp(-lamb * np.sqrt(u))

def c(u):
    return lamb/np.sqrt(u) * np.exp(-lamb * np.sqrt(u))

def d(u):
    return lamb/(2*np.sqrt(u)) * np.exp(-lamb * np.sqrt(u))


print('option ', '(result, uncertainty)')

print('a', integrate.quad(a, 0, 10))
print('b', integrate.quad(b, 0, 10))
print('c', integrate.quad(c, 0, 10))
print('d', integrate.quad(d, 0, 10))
```
